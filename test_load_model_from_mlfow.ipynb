```python
import mlflow
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mlflow.set_tracking_uri("sqlite:///mlflow.db")
# Your model details from the URL
model_name = "catboost_25_12_best"  # From your URL: /models/catboost_25_12_best
model_version = "1"  # From your URL: /versions/1

# Construct the model URI
model_uri = f"models:/{model_name}/{model_version}"
print(f"ðŸ“¥ Loading model: {model_uri}")

# Load the model
model = mlflow.catboost.load_model(model_uri)
print("âœ… Model loaded successfully!")

# Check model type and basic info
print(f"Model type: {type(model)}")
print(f"Model parameters: {model.get_params()}")
```

```python
from utils.index import prep_df


X_train, X_val, y_train, y_val = prep_df(['https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-03.parquet'])

X_full = pd.concat([X_train, X_val])
y_full = pd.concat([y_train, y_val])
y_pred = model.predict(X_full)
rmse = np.sqrt(mean_squared_error(y_full, y_pred))
mae = mean_absolute_error(y_full, y_pred)
r2 = r2_score(y_full, y_pred)

print("\n" + "="*50)
print("MODEL PERFORMANCE ON NEW DATA")
print("="*50)
print(f"RMSE: {rmse:.4f} minutes")
print(f"MAE: {mae:.4f} minutes")
print(f"RÂ² Score: {r2:.4f}")
```
